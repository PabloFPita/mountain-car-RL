```python
import numpy as np
```

```python
# Definir los estados, acciones y la matriz de recompensas R
states = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7']

# Movimientos no permitidos o malos: -100
# Movimientos buenos: 0
# Movimientos buenos con recompensa: 100
# Filas: estados actuales
# Columnas: estados siguientes
R = np.array([
    [-100, 0, 0, -100, -100, -100, -100],
    [-100, -100, 0, -100, -100, -100, -100],
    [-100, 0, -100, -100, 0, -100, -100],
    [-100, 0, 0, -100, 0, 0, -100],
    [-100, -100, 0, -100, -100, 0, 100],
    [-100, -100, -100, -100, 0, -100, 100],
    [-100, -100, -100, -100, 0, 0, 100]
])
print(R)
```

```python
# Inicializar la matriz de calidad Q
Q = np.zeros((7, 7))
Q
```

```python
# Parámetros de Q-Learning
gamma = 0.8 # Factor de descuento (importancia de las recompensas futuras)
alpha = 0.1 # Tasa de aprendizaje (importancia de las recompensas pasadas)
iterations = 1000 # Número de iteraciones
state = 0 # Estado inicial
history = [] # Historial de Q

# Algoritmo Q-Learning
for iteration in range(iterations):
    # Elegir un estado siguiente aleatorio
    next_state = np.random.choice(np.arange(0, 7))
    # Recompensa de ir del estado actual al siguiente
    R[state, next_state]
    # Actualizar Q
    Q[state, next_state] = R[state, next_state] + gamma * np.max(Q[next_state]) # Actualizar Q
    # Pasar al siguiente estado
    state = next_state
    # Mostrar Q cada 100 iteraciones
    if iteration % 100 == 0:
        print(Q)
```

```python
# Ruta óptima
state = 0
while state != 6:
    next_state = np.argmax(Q[state])
    print(f'{states[state]} -> {states[next_state]}')
    state += 1
print(f'{states[state]} -> {states[next_state]}')
```
