```python
import gymnasium as gym
import numpy as np

from stable_baselines3 import TD3
from stable_baselines3.common.noise import NormalActionNoise, OrnsteinUhlenbeckActionNoise

import os
```

```python
# environment
env = gym.make("Pendulum-v1")

env.reset()


# The noise objects for TD3
n_actions = env.action_space.shape[-1]
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))
```

```python
# Saving and loading models 

models_dir = "models/TD3"   # Directory of models for SAC
logdir = "logs"

if not os.path.exists(models_dir):
    os.makedirs(models_dir)

if not os.path.exists(logdir):
    os.makedirs(logdir)
```

```python
# Model creation
model = TD3("MlpPolicy", env, action_noise=action_noise, verbose=1, tensorboard_log=logdir)
```

```python
# Learning

TIMESTEPS = 10000
iters = 0
for i in range(30):
    model.learn(total_timesteps=TIMESTEPS, reset_num_timesteps=False, tb_log_name="TD3")
    # Note the reset_num_timesteps=False. This allows us to see the actual total number of timesteps for the model 
    #rather than resetting every iteration. We're also setting a constant for however many timesteps we want to do 
    #per iteration.
    model.save(f"{models_dir}/{TIMESTEPS*i}")
```

```python
# Testing the results with 10 episodes
env = gym.make("Pendulum-v1")

env.reset()

vec_env = model.get_env()
episodes = 10

for ep in range(episodes):
    #obs = env.reset()
    obs = vec_env.reset()
    done = False
    while not done:
        action, _states = model.predict(obs)
        obs, rewards, dones, info = vec_env.step(action)
        #action, _states = model.predict(obs)
        #obs, rewards, done, info = env.step(action)
        env.render()
        print(rewards)
```
